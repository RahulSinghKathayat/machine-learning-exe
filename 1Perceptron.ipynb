```python
x = modified.iloc[:,:-1]
y = modified["CourseCompletion"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = .20, random_state= 30)
```

```python
from sklearn.linear_model import Perceptron
```

```python
pr = Perceptron(alpha=0.1)
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
# x_train_encoded = lb.fit_transform(x_train)
y_train_encoded = lb.fit_transform(y_train)
# pr.fit(x_train,y_train_encoded)
```

```python
pr.fit(x_train,y_train_encoded)
```

```python
pr.score(x_train,y_train_encoded)
```

```python

pr.score(x_test,y_test)
```
